```python
# B2. Identify and Display the List of following Variables: Index Variables | Categorical Variables - [Nominal | Ordinal] | Non-Categorical, Non-Index Variables.

# Assuming 'Serial_Id' is the index variable (or could be set as index)
index_variables = ['Serial_Id']

# Categorical variables (object dtype)
categorical_variables = ak_60_df_sample.select_dtypes(include='object').columns.tolist()

# Attempt to classify categorical variables into Nominal and Ordinal
# This is a general classification based on common assumptions and may need adjustment
nominal_variables = []
ordinal_variables = []

for col in categorical_variables:
    # Simple heuristic: if there's a clear order (like ship modes), consider it potentially ordinal
    # Otherwise, consider it nominal. This might need manual adjustment based on data understanding.
    if col in ['Ship_Mode']: # Example: Ship_Mode might have an order (Same Day > First Class > Second Class > Standard Class)
        ordinal_variables.append(col)
    else:
        nominal_variables.append(col)

# Non-Categorical, Non-Index Variables (numerical types)
non_categorical_non_index_variables = ak_60_df_sample.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove index variable if it's also a numerical column
if 'Serial_Id' in non_categorical_non_index_variables:
    non_categorical_non_index_variables.remove('Serial_Id')


print("Index Variables:", index_variables)
print("\nCategorical Variables (Nominal):", nominal_variables)
print("\nCategorical Variables (Ordinal - Potential):", ordinal_variables)
print("\nNon-Categorical, Non-Index Variables:", non_categorical_non_index_variables)
```

```python
# B3. Subset (along with Index) and Display the Categorical Variables.

# Combine index and categorical variables
subset_columns = index_variables + nominal_variables + ordinal_variables

# Create the subset DataFrame
ak_60_df_categorical = ak_60_df_sample[subset_columns]

# Display the head of the subset DataFrame
display(ak_60_df_categorical.head())

# Display the shape of the subset DataFrame
print(f"Shape of the categorical subset: {ak_60_df_categorical.shape}")
```

```python
# B4. Subset (along with Index) and Display the Non-Categorical Variables.

# Combine index and non-categorical, non-index variables
subset_columns_non_categorical = index_variables + non_categorical_non_index_variables

# Create the subset DataFrame
ak_60_df_non_categorical = ak_60_df_sample[subset_columns_non_categorical]

# Display the head of the subset DataFrame
display(ak_60_df_non_categorical.head())

# Display the shape of the subset DataFrame
print(f"Shape of the non-categorical subset: {ak_60_df_non_categorical.shape}")
```

```python
# B5. Display the Descriptive Statistics of the Categorical Variables.
# Descriptive Stats: Frequency | Relative-Frequency

categorical_cols = nominal_variables + ordinal_variables

for col in categorical_cols:
    print(f"Descriptive Statistics for '{col}':")
    # Frequency
    frequency = ak_60_df_sample[col].value_counts()
    print("\nFrequency:")
    display(frequency)

    # Relative Frequency
    relative_frequency = ak_60_df_sample[col].value_counts(normalize=True)
    print("\nRelative Frequency:")
    display(relative_frequency)

    print("-" * 30) # Separator for better readability
```

```python
# B6. Visualize the Descriptive Statistics of the Categorical Variables.
# Plots: Bar | Pie (Interactive and Exotic)

import plotly.express as px

categorical_cols = nominal_variables + ordinal_variables

for col in categorical_cols:
    print(f"Visualizing '{col}' with Plotly:")

    # Bar Plot (Frequency)
    # Get value counts and reset index, naming the columns explicitly
    freq_df = ak_60_df_sample[col].value_counts().reset_index()
    freq_df.columns = [col, 'Frequency'] # Rename columns for clarity

    fig_bar = px.bar(
        freq_df,
        x='Frequency',
        y=col,
        orientation='h',
        title=f'Frequency of {col}',
        labels={col: col, 'Frequency': 'Frequency'}
    )
    fig_bar.show()

    # Pie Chart (Relative Frequency)
    fig_pie = px.pie(
        ak_60_df_sample,
        names=col,
        title=f'Relative Frequency of {col}',
        hole=0.3 # Add a hole to make it a donut chart for a slightly more "exotic" look
    )
    fig_pie.update_traces(textposition='inside', textinfo='percent+label')
    fig_pie.show()

    print("-" * 30) # Separator for better readability
```

```python
# C1. Display the Most and Least Preferred Ship_Mode.

# Get the frequency of each Ship_Mode
ship_mode_counts = ak_60_df_sample['Ship_Mode'].value_counts()

# Find the most and least preferred ship modes
most_preferred_ship_mode = ship_mode_counts.idxmax()
least_preferred_ship_mode = ship_mode_counts.idxmin()

print(f"Most Preferred Ship Mode: {most_preferred_ship_mode}")
print(f"Least Preferred Ship Mode: {least_preferred_ship_mode}")
```

```python
# C2. Display Average Delay in Shipment.

# Convert 'Order_Date' and 'Ship_Date' to datetime objects
ak_60_df_sample['Order_Date'] = pd.to_datetime(ak_60_df_sample['Order_Date'], format='%d-%m-%Y')
ak_60_df_sample['Ship_Date'] = pd.to_datetime(ak_60_df_sample['Ship_Date'], format='%d-%m-%Y')

# Calculate the shipment delay
ak_60_df_sample['Ship_Delay'] = ak_60_df_sample['Ship_Date'] - ak_60_df_sample['Order_Date']

# Calculate the average shipment delay
average_ship_delay = ak_60_df_sample['Ship_Delay'].mean()

print(f"Average Shipment Delay: {average_ship_delay}")
```

```python
# C3. Display the Average Variation in Delay in Shipment.

# Calculate the standard deviation of the shipment delay
average_delay_variation = ak_60_df_sample['Ship_Delay'].std()

print(f"Average Variation in Shipment Delay (Standard Deviation): {average_delay_variation}")
```

```python
# C4. Display the Largest and Smallest Business_Segment.

# Get the frequency of each Business_Segment
business_segment_counts = ak_60_df_sample['Business_Segment'].value_counts()

# Find the largest and smallest business segments
largest_business_segment = business_segment_counts.idxmax()
smallest_business_segment = business_segment_counts.idxmin()

print(f"Largest Business Segment: {largest_business_segment}")
print(f"Smallest Business Segment: {smallest_business_segment}")
```

```python
# D1. Create and Display the following Variables:
# 1. Cost [Quantity_Sold * Cost_Per_Unit].
# 2. Revenue [Quantity_Sold * Price_Per_Unit].
# 3. Profit [Revenue - Cost].

ak_60_df_sample['Cost'] = ak_60_df_sample['Quantity_Sold'] * ak_60_df_sample['Cost_Per_Unit']
ak_60_df_sample['Revenue'] = ak_60_df_sample['Quantity_Sold'] * ak_60_df_sample['Price_Per_Unit']
ak_60_df_sample['Profit'] = ak_60_df_sample['Revenue'] - ak_60_df_sample['Cost']

# Display the head of the DataFrame with the new columns
display(ak_60_df_sample.head())
```

```python
# D2. Display the Most and Least Profitable Product_Category.

# Group by Product_Category and calculate the total profit for each category
profit_by_category = ak_60_df_sample.groupby('Product_Category')['Profit'].sum()

# Find the product category with the most and least profit
most_profitable_category = profit_by_category.idxmax()
least_profitable_category = profit_by_category.idxmin()

print(f"Most Profitable Product Category: {most_profitable_category} (Total Profit: {profit_by_category.max():.2f})")
print(f"Least Profitable Product Category: {least_profitable_category} (Total Profit: {profit_by_category.min():.2f})")
```

```python
# D3. Display the Cross-Tabluation of Aggregated Profit in terms of Business_Segment and Product_Category.

# Group by Business_Segment and Product_Category and calculate the sum of Profit
profit_cross_tab = ak_60_df_sample.groupby(['Business_Segment', 'Product_Category'])['Profit'].sum().unstack()

# Display the cross-tabulation
print("Cross-Tabulation of Aggregated Profit by Business Segment and Product Category:")
display(profit_cross_tab)
```

```python
# E1. Filter and Display the Sample Dataset based on: Quantity_Sold < 10 and Profit > Median Value of Profit.

# Calculate the median profit
median_profit = ak_60_df_sample['Profit'].median()

# Filter the dataset based on the conditions
filtered_df = ak_60_df_sample[(ak_60_df_sample['Quantity_Sold'] < 10) & (ak_60_df_sample['Profit'] > median_profit)]

# Display the filtered dataset
print("Filtered Dataset (Quantity_Sold < 10 and Profit > Median Profit):")
display(filtered_df.head())
print(f"Shape of the filtered dataset: {filtered_df.shape}")
```

```python
# E2. Bifurcate and Display the above Filtered Sample Subset into 02 Subsets based on Business_Segment: {Consumer, Corporate}.

# Bifurcate the filtered dataset based on Business_Segment
consumer_subset = filtered_df[filtered_df['Business_Segment'] == 'Consumer']
corporate_subset = filtered_df[filtered_df['Business_Segment'] == 'Corporate']

# Display the head of the Consumer subset
print("Filtered Subset for Consumer Business Segment:")
display(consumer_subset.head())
print(f"Shape of Consumer subset: {consumer_subset.shape}")

print("\nFiltered Subset for Corporate Business Segment:")
display(corporate_subset.head())
print(f"Shape of Corporate subset: {corporate_subset.shape}")
```

```python
# E3. Sort and Display both the Bifurcated Business_Segments Subsets: {Consumer, Corporate}, based on {Product_Category: Ascending, Quantity_Sold: Ascending, Profit: Descending}.

# Sort the Consumer subset
consumer_subset_sorted = consumer_subset.sort_values(by=['Product_Category', 'Quantity_Sold', 'Profit'],
                                                   ascending=[True, True, False])

# Sort the Corporate subset
corporate_subset_sorted = corporate_subset.sort_values(by=['Product_Category', 'Quantity_Sold', 'Profit'],
                                                      ascending=[True, True, False])

# Display the head of the sorted Consumer subset
print("Sorted Consumer Business Segment Subset:")
display(consumer_subset_sorted.head())

print("\nSorted Corporate Business Segment Subset:")
display(corporate_subset_sorted.head())
```

```python
# E4. Append and Display the Sorted Business_Segments Subsets: Corporate followed by Consumer.

# Append the sorted Corporate subset followed by the sorted Consumer subset
appended_subsets = pd.concat([corporate_subset_sorted, consumer_subset_sorted])

# Display the head and tail of the appended DataFrame to show the order
print("Appended Business Segments Subsets (Corporate followed by Consumer):")
display(appended_subsets.head())
display(appended_subsets.tail())

print(f"\nShape of the appended dataset: {appended_subsets.shape}")
```

```python
# E5. Merge and Display the Filtered Business_Segment Subsets: {Consumer, Corporate} in the following Manner:
# 1. For Business_Segment = Consumer, Consider only 02 Variables: {Order_Date, Region}.
# 2. For Business_Segment = Corporate, Consider only 02 Variables: {Order_Date, Product_Category}.
# 3. Merge the 02 Business_Segment Subsets on the Order_Date.

# Select specific columns from the Consumer subset
consumer_subset_selected = consumer_subset[['Order_Date', 'Region']]

# Select specific columns from the Corporate subset
corporate_subset_selected = corporate_subset[['Order_Date', 'Product_Category']]

# Merge the two subsets on 'Order_Date'
# Using an outer merge to keep all dates from both subsets
merged_subsets = pd.merge(consumer_subset_selected, corporate_subset_selected, on='Order_Date', how='outer')

# Display the merged dataset
print("Merged Subsets (Consumer with Region, Corporate with Product_Category, merged on Order_Date):")
display(merged_subsets.head())
print(f"Shape of the merged dataset: {merged_subsets.shape}")
```

```python
# E6. Display the Region-wise Count and Product_Category-wise Count from the Merged Subset.

print("Region-wise Count from Merged Subset:")
display(merged_subsets['Region'].value_counts())

print("\nProduct_Category-wise Count from Merged Subset:")
display(merged_subsets['Product_Category'].value_counts())
```
